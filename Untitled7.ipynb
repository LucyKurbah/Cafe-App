```python
import pytesseract
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt


# Set the path to Tesseract OCR
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'



# Call the function to get line item coordinates


def mark_region(image_path):
    im = cv2.imread(image_path)
    gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
  # closes all windows opened by OpenCV
    
    blur = cv2.GaussianBlur(gray, (9,9), 0)
    thresh = cv2.adaptiveThreshold(blur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV,11,5)

    # Dilate to combine adjacent text contours
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (9,9))
    dilate = cv2.dilate(thresh, kernel, iterations=4)
    # cv2_imshow(thresh)
    # cv2.waitKey(0)  # waits for any key to be pressed
    # cv2.destroyAllWindows()

    # Find contours, highlight text areas, and extract ROIs
    cnts = cv2.findContours(dilate, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cnts = cnts[0] if len(cnts) == 2 else cnts[1]
    # cv2_imshow(dilate)
    # cv2.waitKey(0)  # waits for any key to be pressed
    # cv2.destroyAllWindows()

    if not cnts:
      return None, None
    line_items_coordinates = []
    image = im
    # cv2_imshow(image)
    # cv2.waitKey(0)  # waits for any key to be pressed
    # cv2.destroyAllWindows()

    for c in cnts:
        area = cv2.contourArea(c)

        x,y,w,h = cv2.boundingRect(c)

        if y >= 600 and x <= 1000:
            if area > 10000:
                image = cv2.rectangle(im, (x,y), (w+x, y+h), color=(255,0,255), thickness=3)
                line_items_coordinates.append([(x,y), (w+x, y+h)])

        if y >= 2400 and x<= 2000:
            image = cv2.rectangle(im, (x,y), (w+x, y+h), color=(255,0,255), thickness=3)
            line_items_coordinates.append([(x,y), (w+x, y+h)])
            # if area > 10000:
            #     line_items_coordinates.append([(x,y), (w+x, y+h)])

    # print(line_items_coordinates)
    return image, line_items_coordinates
# Load the original image
image = cv2.imread('one.jpg')

annotated_image, line_items_coordinates = mark_region('/content/Page_2.jpg')
# Get co-ordinates to crop the image
cv2_imshow(annotated_image)
cv2.waitKey(0)  # waits for any key to be pressed
cv2.destroyAllWindows()

c = line_items_coordinates[1]
print(c)
# Cropping image img = image[y0:y1, x0:x1]
img = annotated_image[c[0][1]:c[1][1], c[0][0]:c[1][0]]

plt.figure(figsize=(1,1))
plt.imshow(img)

# Convert the image to black and white for better OCR
ret, thresh1 = cv2.threshold(annotated_image, 120, 255, cv2.THRESH_BINARY)

# Pytesseract image to string to get results
text = str(pytesseract.image_to_string(thresh1, config='--psm 6'))
print(text)
```
